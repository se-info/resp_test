```python
train_df = pd.read_csv('/content/sample_data/california_housing_train.csv')
x_train = train_df.drop('median_house_value', axis = 1)
y_train = train_df['median_house_value']
```

```python
X = np.concatenate((x_train, np.ones((x_train.shape[0], 1))), axis=1)
Y = np.reshape(y_train.to_numpy(), (y_train.shape[0], 1))
X.shape,Y.shape
```

```python
total_rows = len(train_df)
train_rows = int(0.8 * total_rows)
# test_rows = total_rows - train_rows

trainX, testX = X[:train_rows], X[train_rows:] # dev/validation
trainY, testY = Y[:train_rows], Y[train_rows:]
trainX.shape,testX.shape
```

```python
def normalize_data(train_x, test_x):
    train_mean = np.mean(train_x, axis=(0,1), dtype=np.float64, keepdims=True)
    train_std = np.std(train_x, axis=(0,1), dtype=np.float64, keepdims=True)
    train_x = (train_x-train_mean)/train_std
    test_x = (test_x-train_mean)/train_std
    return train_x, test_x
```

```python
trainX, testX = normalize_data(trainX, testX)
```

```python
model = Sequential()
model.add(Dense(1, activation='sigmoid', input_shape=(2,)))
model.compile(SGD(0.001), loss='binary_crossentropy', metrics=['acc'])
model.fit(trainX, trainY, epochs=200, validation_data=(testX, testY))
```
