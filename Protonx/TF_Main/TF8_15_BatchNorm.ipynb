```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical

# Load and preprocess the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0  # Normalize the data
y_train, y_test = to_categorical(y_train, 10), to_categorical(y_test, 10)  # One-hot encode labels

# Create a Sequential model without batch normalization
model_without_bn = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model_without_bn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model_without_bn.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.2)

# Evaluate the model
test_loss_without_bn, test_acc_without_bn = model_without_bn.evaluate(x_test, y_test)
print(f"Test Accuracy without Batch Normalization: {test_acc_without_bn}")
```

```python
from tensorflow.keras.layers import BatchNormalization

# Create a Sequential model with batch normalization
model_with_bn = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    BatchNormalization(),
    # tf.keras.layers.Activation('relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model_with_bn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model_with_bn.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model
test_loss_with_bn, test_acc_with_bn = model_with_bn.evaluate(x_test, y_test)
print(f"Test Accuracy with Batch Normalization: {test_acc_with_bn}")
```
