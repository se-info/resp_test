```python
y = np.array([
    [13],
    [35],
    [41],
    [19],
    [45],
    [28],
    [10],
    [55]
])
```

```python
# theta = np.random.normal(size = 2).reshape(2,1)
theta = np.array([[1],[1]])
theta
```

```python
def hypo(X,theta):
  return X.dot(theta)
```

**Lập trình Gradient**

```python
def gradient(X,y,theta,m):
  gradient = 1/m * X.T.dot(hypo(X,theta) - y) # (2 x 1)
  return gradient
  #Linear Regression Model
```

```python
def cost(X, y, theta, m):
  # Lập trình tại đây
  v = sum((hypo(X,theta)-y)**2)/(2*m)
  return v
```

```python
# cost(X, y, theta, m)
```

```python
# hyperparameters
# fine tune
m = 8
learning_rate = 0.001
epochs = 300
```

```python
# gradient(X,y,theta,m)
```

```python
for i in range(epochs):
  # Tính gradient
  grad = gradient(X,y,theta,m)
  # print(grad)
  theta = theta - learning_rate * grad
  if (i+1) % 3 == 0:
    print('Cost is: {}'.format(cost(X,y,theta,m)))
```
