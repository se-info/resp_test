## 2. Giải quyết

### 2.1 Đạo hàm của $a$ trên $z$

<img src="https://storage.googleapis.com/protonx-cloud-storage/images/Backprop8.PNG" >



Để đơn giản công thức thì trong chứng minh này, sẽ chứng minh công thức trên 1 điểm dữ liệu.

$$
\textbf{e} = \nabla_{\textbf{z}}J = \hat{\textbf{y}} - \textbf{y}
$$

Giả sử $c = 10$

$$
  a_1 = \frac{e^{z_1}}{e^{z_1} + e^{z_2} + \dots + e^{z_{10}}}
$$

Đạo hàm của $a_1$ trên $z_1$:

$$
\frac{\partial{a_1}}{\partial{z_1}} = \frac{\partial{\frac{e^{z_1}}{e^{z_1} + e^{z_2} + \dots + e^{z_{10}}}}}{\partial{z_1}}
$$

Áp dụng công thức nhân + quy tắc chuỗi

$$
\frac{\partial{a_1}}{\partial{z_1}} = \frac{\partial{e^{z_1}}}{\partial{z_1}}\frac{1}{e^{z_1} + e^{z_2} + \dots + e^{z_{10}}} + e^{z_1} \frac{\partial{\frac{1}{e^{z_1} + e^{z_2} + \dots + e^{z_{10}}}}}{\partial{(e^{z_1} + e^{z_2} + \dots + e^{z_{10}})}} \frac{\partial{(e^{z_1} + e^{z_2} + \dots + e^{z_{10}})}}{\partial{e^{z_1}}} \frac{\partial{e^{z_1}}}{\partial{z_1}} \\
= \frac{e^{z_1}}{e^{z_1} + e^{z_2} + \dots + e^{z_{10}}} - e^{z_1} \frac{1}{(e^{z_1} + e^{z_2} + \dots + e^{z_{10}})^2}e^{z_1}\\
= \frac{e^{z_1}}{(e^{z_1} + e^{z_2} + \dots + e^{z_{10}})}\begin{pmatrix}
& 1 - \frac{e^{z_1}}{(e^{z_1} + e^{z_2} + \dots + e^{z_{10}})} \end{pmatrix} = a_1 (1 - a_1) \tag{1}
$$

Đạo hàm của $a_1$ trên $z_2$:

$$
\frac{\partial{a_1}}{\partial{z_2}} = \frac{\partial{\frac{e^{z_1}}{e^{z_1} + e^{z_2} + \dots + e^{z_{10}}}}}{\partial{z_2}} \tag{2}
$$


$$
\frac{\partial{a_1}}{\partial{z_2}} = \frac{\partial{e^{z_1}}}{\partial{z_2}}\frac{1}{e^{z_1} + e^{z_2} + \dots + e^{z_{10}}} + e^{z_1} \frac{\partial{\frac{1}{e^{z_1} + e^{z_2} + \dots + e^{z_{10}}}}}{\partial{(e^{z_1} + e^{z_2} + \dots + e^{z_{10}})}} \frac{\partial{(e^{z_1} + e^{z_2} + \dots + e^{z_{10}})}}{\partial{e^{z_2}}} \\
= \frac{- e^{z_1}e^{z_2}}{(e^{z_1} + e^{z_2} + \dots + e^{z_{10}})} = - \frac{e^{z_1}}{(e^{z_1} + e^{z_2} + \dots + e^{z_{10}})} \frac{e^{z_2}}{(e^{z_1} + e^{z_2} + \dots + e^{z_{10}})} = -a_1a_2
$$


Tương tự:

$$
\frac{\partial{a_1}}{\partial{z_3}}
= \frac{- e^{z_1}e^{z_3}}{(e^{z_1} + e^{z_2} + \dots + e^{z_{10}})^2} = -a_1a_3 \tag{3}
$$

$$
\vdots
$$

$$
\frac{\partial{a_1}}{\partial{z_{10}}}
= \frac{- e^{z_1}e^{z_{10}}}{(e^{z_1} + e^{z_2} + \dots + e^{z_{10}})^2} = -a_1a_{10} \tag{4}
$$



Công thức này có thể viết gọn dưới dạng như sau:

$$
\frac{\partial{a_i}}{\partial{z_j}} = \left\{\begin{matrix}
a_i(1-a_i) \ \ \forall i = j \\
-a_ia_j \ \ \forall i \neq j\\
\end{matrix}\right.
$$

### 2.2 - Đạo hàm của $J$ trên $\textbf{z}$



Công thức toán của cross-entropy cost $J$: $$J(\textbf{W}) = -\frac{1}{m} \sum\limits_{i = 1}^{m} \sum\limits_{j = 1}^{c} \textbf{y}^{(i)}\log (\hat
{\textbf{y}}^{(i)})$$

Với
- $c$: số nhãn
- $m$: số điểm dữ liệu

Loss Function trên 1 điểm dữ liệu

$$J(\textbf{W}) = - \sum\limits_{j = 1}^{c} \textbf{y}\log (\hat
{\textbf{y}})$$

Với 10 nhãn thì

$$J(\textbf{W}) = - \sum\limits_{j = 1}^{10} \textbf{y}\log (\hat
{\textbf{y}})$$


Giả sử nhãn

$$\textbf{y} = \begin{bmatrix}
1
\\ 0
\\ \vdots
\\ 0
\end{bmatrix}
$$

và đầu ra của hàm softmax:

$$\hat{\textbf{y}} = \begin{bmatrix}
a_1
\\ a_2
\\ \vdots
\\
\\ a_{10}
\end{bmatrix}
$$

Cho nên $$J(\textbf{W}) = - \sum\limits \begin{pmatrix}
\begin{bmatrix}
1
\\ 0
\\ \vdots
\\ 0
\end{bmatrix}
 & \odot  \log\begin{pmatrix}\begin{bmatrix}
a_1
\\ a_2
\\ \vdots
\\
\\ a_{10}
\end{bmatrix} \end{pmatrix}
\end{pmatrix}
= - \sum\limits  \begin{pmatrix}
 \begin{bmatrix}
\log{(a_1)}
\\ 0
\\ \vdots
\\ 0
\end{bmatrix}
\end{pmatrix} = - \log{(a_1)}$$

Cho nên

$$
 \frac{\partial{J}}{\partial{a_1}} = - \frac{\partial{\log{(a_1)}}}{\partial{a_1}} = -\frac{1}{a_1}
$$

$$
 \frac{\partial{J}}{\partial{a_2}} = - \frac{\partial{\log{(a_1)}}}{\partial{a_2}} = 0
$$

$$
\vdots
$$

$$
 \frac{\partial{J}}{\partial{a_{10}}} = - \frac{\partial{\log{(a_{1})}}}{\partial{a_{10}}} = 0
$$

Đạo hàm của $J$ trên $\textbf{z}$

Đạo hàm của $J$ trên $z_1$

Vì $z_1$ ảnh hưởng đến $J$ thông qua các giá trị $a$ cho nên sử dụng công thức quy tắc chuỗi nhiều biến ta có:

$$
\frac{\partial{J}}{\partial{z_1}} = \sum\limits_{j = 1}^{10} \frac{\partial{J}}{\partial{a_j}} \frac{\partial{a_j}}{\partial{z_1}} \\
= \frac{\partial{J}}{\partial{a_1}} \frac{\partial{a_1}}{\partial{z_1}} + \frac{\partial{J}}{\partial{a_2}} \frac{\partial{a_2}}{\partial{z_1}} + \dots + \frac{\partial{J}}{\partial{a_{10}}} \frac{\partial{a_{10}}}{\partial{z_1}} \\
= \frac{\partial{J}}{\partial{a_1}} \frac{\partial{a_1}}{\partial{z_1}} \\
$$

Kết hợp công thức số $(1):$

$$
\frac{\partial{J}}{\partial{z_1}} = -\frac{1}{a_1} a_1 (1 - a_1) = a_1 - 1
$$

Đạo hàm của $J$ trên $z_2$

$$
\frac{\partial{J}}{\partial{z_2}} = \sum\limits_{j = 1}^{10} \frac{\partial{J}}{\partial{a_j}} \frac{\partial{a_j}}{\partial{z_2}} \\
= \frac{\partial{J}}{\partial{a_1}} \frac{\partial{a_1}}{\partial{z_2}} + \frac{\partial{J}}{\partial{a_2}} \frac{\partial{a_2}}{\partial{z_2}} + \dots + \frac{\partial{J}}{\partial{a_{10}}} \frac{\partial{a_{10}}}{\partial{z_2}} \\
=  \frac{\partial{J}}{\partial{a_1}} \frac{\partial{a_1}}{\partial{z_2}} \\
$$

Kết hợp công thức số $(2):$

$$\frac{\partial{J}}{\partial{z_2}} = -\frac{1}{a_1} (-a_1a_2) = a_2$$

Tương tự

Đạo hàm của $J$ trên $z_3$:
$$
\frac{\partial{J}}{\partial{z_3}} = a_3
$$

$$
\vdots
$$

$$
\frac{\partial{J}}{\partial{z_{10}}} = a_{10}
$$


Suy ra Gradient của hàm $J$ trên vector $\textbf{z}$ sẽ là:

Trường hợp nay ta để chiều Gradient giống chiều của $\textbf{y}$ và $\hat{\textbf{y}}$ để dễ nhìn thấy quan hệ.

$$
\textbf{e} = \nabla_{\textbf{z}}J = \begin{bmatrix}
\frac{\partial{J}}{\partial{z_1}}  \\  
\frac{\partial{J}}{\partial{z_2}} \\  
\frac{\partial{J}}{\partial{z_3}}  \\  
\dots \\  \frac{\partial{J}}{\partial{z_{10}}}
\end{bmatrix} = \begin{bmatrix}
a_1 - 1
\\ a_2
\\ a_3 \\  
\vdots \\
a_{10}\end{bmatrix}
= \begin{bmatrix}
a_1  \\
a_2 \\
a_3 \\  
\vdots \\
a_{10}
\end{bmatrix} - \begin{bmatrix}
1  \\
0 \\
0 \\  
\vdots \\
0
\end{bmatrix}
= \hat{\textbf{y}} - \textbf{y}
$$

**Kết luận:** Từ đây có thể suy ra công thức trên toàn bộ tập dữ liệu.

$$ \textbf{E}^{(L)} = \hat{\textbf{Y}} - \textbf{Y} $$
