```python
# Import Relevant libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Learning rate
learning_rate = 0.01

# Number of loops for training through all your data to update the parameters
training_epochs = 100

# the training dataset
x_train = np.linspace(0, 10, 100)
y_train = x_train + np.random.normal(0,1,100)

# plot of data
plt.scatter(x_train, y_train)
```

```python
# declare weights
a = tf.Variable(0.)
b = tf.Variable(0.)
```

```python
# Define linear regression expression y
def linreg(x):
    y = a*x + b
    return y
```

```python
# Define loss function (MSE)
def squared_error(y_pred, y_true):
    return tf.reduce_mean(tf.square(y_pred - y_true))
```

```python
y_predicted = linreg(x_train)
loss = squared_error(y_predicted, y_train)
[a,b]
```

```python
# train model
for epoch in range(training_epochs):
    # Compute loss within Gradient Tape context
    with tf.GradientTape(persistent=True) as tape:
        y_predicted = linreg(x_train)
        loss = squared_error(y_predicted, y_train)

        # Get gradients
        gradients = tape.gradient(loss, [a,b])

        # Adjust weights
        # a = g_loss_a*learning_rate
        # b = g_loss_b*learning_rate
        a.assign_sub(gradients[0]*learning_rate)
        b.assign_sub(gradients[1]*learning_rate)

        # Print output
        print(f"Epoch count {epoch}: Loss value: {loss.numpy()}")
```

```python
import os
# Set environment variables
os.environ['PROTOCOL_BUFFERS_PYTHON_IMPLEMENTATION'] = 'python'
import tensorflow as tf

@tf.function
def exercise07(W, x):
  """
  Đầu vào:
    - W là ma trận
    - x là vector
  Đầu ra:
    - Jacobian của f với W và x
  Lưu ý: sử dụng GradientTape.
  """

  # Lập trình tại đây
  # TODO 1. Chuyển số thực thành tf.constant
  W = tf.constant(W,dtype=tf.float32)
  x = tf.constant(x,dtype=tf.float32)
  # TODO 2. Khởi tạo Gradient Tape.
  with tf.GradientTape(persistent=True) as tape:
    tape.watch(W)
    tape.watch(x)

    # TODO 3. Lập trình công thức hàm f
    f = tf.matmul(W,x)
    print(f)

  # TODO 4. Tính toán đạo hàm
  jfw = tape.jacobian(f,W)
  jfx = tape.jacobian(f,x)
  result = [jfw, jfx]
  print(jfw,jfx)
  return result

try:
  W = [[1.0, 2.0, 3.0], [-1.0, 3.0, 2.0], [2.0, 3.0, 1.0]]
  x = [[4.0], [5.0], [6.0]]
  result = exercise07(W, x)
  # print(result[0])
  print("Jacobian của f với W: {}".format(result[0]))
  print("Jacobian của f với x: {}".format(result[1]))
except Exception as e:
  print(e)
```

```python
import os
import numpy as np
# Set environment variables
os.environ['PROTOCOL_BUFFERS_PYTHON_IMPLEMENTATION'] = 'python'
import tensorflow as tf
@tf.function
def exercise08(A, x):
  """
  Đầu vào:
    - A là ma trận
    - x là vector
  Đầu ra:
    - Jacobian của y với A và x
  Lưu ý: sử dụng GradientTape.
  """

  # Lập trình tại đây
  # TODO 1. Chuyển số thực thành tf.constant
  A = tf.constant(A,dtype=tf.float32)
  x = tf.constant(x,dtype=tf.float32)
  # TODO 2. Khởi tạo Gradient Tape.
  with tf.GradientTape(persistent=True) as tape:
    tape.watch(A)
    tape.watch(x)

    # TODO 3. Lập trình công thức hàm f
    f = tf.matmul(tf.matmul(x, A, transpose_a=True),x)
    print(f)

  # TODO 4. Tính toán đạo hàm
  jfa = tape.gradient(f,A)
  jfx = tape.gradient(f,x)
  # print(jfa,jfx)
  result = [jfa, jfx]
  return f

try:
  A = [[1.0, 10.0, 3.0], [-1.0, 30.0, 2.0], [12.0, 3.0, 1.0]]
  x = [[4.0], [20.0], [1.0]]
  result = exercise08(A,x)
  print("Jacobian của f với A: {}".format(result[0]))
  print("Jacobian của f với x: {}".format(result[1]))
except Exception as e:
  print(e)
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt


# train model
# ...

a = tf.Variable(0.)
b = tf.Variable(0.)
training_epochs = 100
learning_rate = 0.01

for epoch in range(training_epochs):
    fig, ax = plt.subplots()
    ax.scatter(x_train, y_train, label='Data Points')

    # Initialize the line with a and b
    line, = ax.plot(x_train, a * x_train + b, label='Fitted Line', color='red')
    # Compute loss within Gradient Tape context
    with tf.GradientTape(persistent=True) as tape:
        tape.watch(a)
        tape.watch(b)
        y_predicted = linreg(x_train)
        loss = squared_error(y_predicted, y_train)

        # Get gradients
        g_loss_a = tape.gradient(loss, a)
        g_loss_b = tape.gradient(loss, b)

        # Adjust weights
        a = a- g_loss_a * learning_rate
        b = b- g_loss_b * learning_rate

        # Update the line on the plot
        line.set_ydata(a * x_train + b)

        # Print output
        print(f"Epoch count {epoch}: Loss value: {loss.numpy()}")

        # Update the plot
        plt.pause(0.01)

plt.show()
```
